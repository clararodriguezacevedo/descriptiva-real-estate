# Cargamos los tsv:

```python
zonaprop_alquiler = pd.read_csv('../data/raw/zonaprop_alquiler_1776024151.tsv', sep = '\t')

zonaprop_alquiler.shape
zonaprop_alquiler.head()
```

```python
zonaprop_alquiler_temporal = pd.read_csv('../data/raw/zonaprop_alquiler_temporal_1776025830.tsv', sep = '\t')

zonaprop_alquiler_temporal.shape
zonaprop_alquiler_temporal.head()
```

```python
zonaprop_venta = pd.read_csv('../data/raw/zonaprop_venta_1776027996.tsv', sep = '\t')

zonaprop_venta.shape
zonaprop_venta.head()
```

```python
argenprop_alquiler = pd.read_csv('../data/raw/argenprop_alquiler.tsv', sep = '\t')

argenprop_alquiler.shape
argenprop_alquiler.head()
```

```python
argenprop_alquiler_temporal = pd.read_csv('../data/raw/argenprop_alquiler_temporal.tsv', sep = '\t')

argenprop_alquiler_temporal.shape
argenprop_alquiler_temporal.head()
```

```python
argenprop_venta = pd.read_csv('../data/raw/argenprop_venta.tsv', sep = '\t')

argenprop_venta.shape
argenprop_venta.head()
```

# Concatenamos todos los datos recolectados:

Como desarrollamos los scrapers para que todas las tablas tengan las mismas columnas, con el mismo formato, podemos concatenar directamente:

```python
df_real_estate = pd.concat([zonaprop_alquiler,zonaprop_alquiler_temporal,zonaprop_venta,argenprop_alquiler,argenprop_alquiler_temporal,argenprop_venta])
df_real_estate.shape
```

```python
df_real_estate.head()
```

# Descargar el dataframe maestro:

```python
import os

OUTPUT_DIR = "../data/raw"
os.makedirs(OUTPUT_DIR, exist_ok=True)

filename = "dataframe_maestro.tsv"
filepath = os.path.join(OUTPUT_DIR, filename)
df_real_estate.to_csv(filepath, sep='\t', index=False, encoding='utf-8-sig')
print(f"\n✅ ¡Éxito! Guardado en: {filepath}")
```
